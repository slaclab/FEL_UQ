```python
def get_scale(data,lower=-1,upper=1):
    offset =np.nanmin(data,axis=0)
    scale=np.nanmax(data,axis=0)-np.nanmin(data,axis=0)
    return offset, scale
def scale_data(data,offset,scale,lower=-1,upper=1):
    data_scaled=lower + ((data-offset)*(upper-lower)/scale)
    return data_scaled
def unscale_data(data,offset,scale,lower=-1,upper=1):
    data_unscaled=(((data-lower)*scale)/(upper-lower)) + offset
    return data_unscaled
```

```python
# define class for showing training plot - found online
class PlotLosses(keras.callbacks.Callback):
    def on_train_begin(self, logs={}):
        self.i = 0
        self.x = []
        self.losses = []
        self.val_losses = []
        
        self.fig = plt.figure()
        
        self.logs = []

    def on_epoch_end(self, epoch, logs={}):
        
        self.logs.append(logs)
        self.x.append(self.i)
        self.losses.append(logs.get('loss'))
        self.val_losses.append(logs.get('val_loss'))
        self.i += 1
        
        clear_output(wait=True)
        plt.plot(self.x, self.losses, label="loss")
        plt.plot(self.x, self.val_losses, label="val_loss")
        plt.legend()
        plt.ylabel('error')
        plt.xlabel('epoch')
        plt.show();
        
        
        
# define class for showing training plot - found online
class PlotLogLosses(keras.callbacks.Callback):
    def on_train_begin(self, logs={}):
        self.i = 0
        self.x = []
        self.losses = []
        self.val_losses = []
        
        self.fig = plt.figure()
        
        self.logs = []

    def on_epoch_end(self, epoch, logs={}):
        
        self.logs.append(logs)
        self.x.append(self.i)
        self.losses.append(logs.get('loss'))
        self.val_losses.append(logs.get('val_loss'))
        self.i += 1
        
        clear_output(wait=True)
        plt.yscale('log')
        plt.plot(self.x, self.losses, label="loss")
        plt.plot(self.x, self.val_losses, label="val_loss")
 
        plt.legend()
        plt.ylabel('error')
        plt.xlabel('epoch')
        plt.show();
        
plot_losses = PlotLogLosses()
```

```python
def build_ff_mdl(in_dim = 4, out_dim = 2, l1 = 8, l2 = 6, l3 = 4, l4= 4,l5=5,l6=2, opt = 'adam', loss = 'mse', act = 'tanh', l2reg = 0.00001):
    layer_0 = Input(shape=(in_dim,))
    layer_1 = Dense(l1, activation=act, kernel_regularizer=regularizers.l2(l2reg))(layer_0)
    ld=GaussianNoise(0.1)(layer_1)
    layer_2 = Dense(l2, activation=act, kernel_regularizer=regularizers.l2(l2reg))(ld)
    ld=GaussianNoise(0.1)(layer_2)
    layer_3 = Dense(l3, activation=act, kernel_regularizer=regularizers.l2(l2reg))(ld)
    ld=GaussianNoise(0.1)(layer_3)
    layer_4 = Dense(l4, activation=act, kernel_regularizer=regularizers.l2(l2reg))(ld)
    ld=GaussianNoise(0.1)(layer_4)
    layer_4 = Dense(l5, activation=act, kernel_regularizer=regularizers.l2(l2reg))(ld)
    layer_4 = Dense(l6, activation=act, kernel_regularizer=regularizers.l2(l2reg))(layer_4)
  
    layer_5 = Dense(out_dim, activation='linear')(layer_4)
    model = Model(inputs=layer_0, outputs=layer_5)
    model.compile(loss=loss, optimizer=opt, metrics=['mse'])
    return model
```
